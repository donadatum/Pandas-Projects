## Collecting Twitter Data

### Twitter has an API to enable collection of streaming Twitter data.  The API documentation is extensive.  Keys can be obtained by logging in to apps.twitter.com. 

#### The script below uses tweepy to collect and print 1,000 tweets with the #Delta handle.  Alternatively tweets can be saved to a JSON file.

```python
import tweepy
import pprint
import sys
from tweepy import OAuthHandler

consumer_key = 'XXXXXXXXXXX'
consumer_secret = 'XXXXXXXXXXX'
access_token = 'XXXXXXXXX'
access_secret = 'XXXXXXXXXX'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
deltaTweet = tweepy.Cursor(api.search, q='#Delta').items(1000)

for tweet in deltaTweet:
   print (tweet.created_at, tweet.text)
   #results = tweet.created_at + "," +  tweet_text
   with open('delta2.txt','a') as f:
       f.write(tweet.text)
```
